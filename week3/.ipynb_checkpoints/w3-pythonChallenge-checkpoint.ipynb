```python
#python challenge 3
# http://www.pythonchallenge.com/pc/def/equality.html
# again, in the source code, we can see the chracters.
# the hint is to find any one small letter, surrounded by EXACTLY three big bodyguards on each of its sides.

# use the RegEx Module, the built in python package.
#https://www.w3schools.com/python/python_regex.asp
import re

#save the mess characters into a txt file, and read it.
data = open('challenge-3.txt').read()
# replace the change line symbol to empty strings.
newData = data.replace("\n", "")

# find the characters
#[^A-Z]+ means any characters except capital characters; + means one or more occurances. 
# Because we want EXACTLY three big bodyguards, so BEFORE and AFTER the three big ones, it must be small chracters.
# [A-Z]{3} means three occurances of capital characters.
# ([a-z]) groups the center small letter. only one.
newData = re.findall("[^A-Z]+[A-Z]{3}([a-z])[A-Z]{3}[^A-Z]+", data)
result = "".join(newData)
urlResult = "".join(("http://www.pythonchallenge.com/pc/def/", result, ".html"))
print(urlResult)
```

```python
#python challenge 6
#http://www.pythonchallenge.com/pc/def/channel.html
#In the source code, at the top line besides <html> is <!-- <-- zip -->, page title is "now there are pairs"

#so change the link to http://www.pythonchallenge.com/pc/def/channel.zip
#download it and unzip it. There are many txt files with one readme.
#in the readme.txt, it says "welcome to my zipped list. hint1: start from 90052 hint2: answer is inside the zip"

#In the txt files, the contents of "the next nothing is..." brought me back to the linked list challenge.

#tutorial working with zip:https://www.geeksforgeeks.org/working-zip-files-python/
import zipfile, re
f = zipfile.ZipFile("challenge6-channel.zip")
num = '90052'
comments = []
while True:
    content = f.read(num + ".txt").decode("utf-8")
    #print(content)
    c = f.getinfo(num + ".txt").comment.decode("utf-8")
    comments.append(c)
    match = re.search("Next nothing is (\d+)", content)
    if match == None:
        break
    num = match.group(1)
#At the end of the loop, we got hint: Collect the comments.
#About getting comment, we use getinfo() and comment; 
#related info:https://stackoverflow.com/questions/51792923/python-reading-a-zip-file-comment
#So I went back to the while loop and add comment

print("".join(comments))
#After trying hockey.html, it links me to observe the chracters in the banner. 
#so it's oxygen
```

```python
#python challenge 7
# http://www.pythonchallenge.com/pc/def/oxygen.html
# first install pillow in terminal: $ pip3 install pillow

# load image from URL
import requests
from io import BytesIO
# load the Image module from pillow
from PIL import Image
img = Image.open(BytesIO(requests.get('http://www.pythonchallenge.com/pc/def/oxygen.png').content))

# print("width is {}, height is {}".format(img.width,img.height))
#get pixel on RGBA value
# print(img.getpixel((0,0)))

#get pixel value on the middle row
row = [img.getpixel((x, img.height / 2)) for x in range(img.width)]

#we only need the greyScale numbers.
#row2 collects the unique RGBA values
#row3 collects the greyscale value. since r = g = b
row2 = [row[0]]
for x in row:
    if x != row2[-1]:
        row2.append(x)
row3 = []
for x in row2:
    r = x[0]
    g = x[1]
    b = x[2]
    if ( r == b == g):
        row3.append(x[0])
print(row3)
#map the greyscale numbers with the ASCII characters.
output  = "".join(map(chr, row3))
print(output)

#There is another way to calculate row3. I learned it from the python challenge solution. it used slicing method to quickly get data. since rach block is 7 pixel long.
row = row[::7]
output = []
#get only the r value.
for r,g,b,a in row:
    if r ==g ==b:
        output.append(r)

#I got a new hint: smart guy, you made it. the next level is [105, 10, 16, 101, 103, 14, 105, 16, 121]
#I used re findall method again. \d means returning a match where the string contains digits (numbers from 0-9)
nums = re.findall("\d+", "".join(map(chr, output)))
print(nums)
result = "".join(map(chr, map(int, nums)))
print(result)
#so the result is "integrity"
```
